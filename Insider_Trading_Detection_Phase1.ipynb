```python
owner_sample = pd.read_csv(
    'lit_reportingowner.csv',
    nrows=200_000
)

owner_sample.head()
```

```python
owner_sample.to_csv('owner_sample.csv', index=False)
```

```python
important_cols = [
    'accessionNumber',
    'filerCik',
    'filingDate',
    'transactionDate',
    'transactionAcquiredDisposedCode',
    'transactionShares',
    'sharesOwnedFollowingTransaction',
    'directOrIndirectOwnership'
]

nonderiv_clean = sample_df[important_cols]
nonderiv_clean.head()
```

```python
nonderiv_clean = nonderiv_clean.rename(columns={
    'transactionAcquiredDisposedCode': 'buy_sell',
    'transactionShares': 'shares',
    'sharesOwnedFollowingTransaction': 'shares_after'
})
```

```python
nonderiv_clean['buy_sell'] = nonderiv_clean['buy_sell'].map({
    'A': 1,  # Buy (Acquire)
    'D': 0   # Sell (Dispose)
})
```

```python
nonderiv_clean = nonderiv_clean.dropna(
    subset=['buy_sell', 'shares', 'transactionDate']
)
```

```python
nonderiv_clean['transactionDate'] = pd.to_datetime(
    nonderiv_clean['transactionDate'], errors='coerce'
)
```

```python
nonderiv_clean.info()
```

```python
nonderiv_clean['shares'] = pd.to_numeric(
    nonderiv_clean['shares'], errors='coerce'
)

nonderiv_clean['shares_after'] = pd.to_numeric(
    nonderiv_clean['shares_after'], errors='coerce'
)
```

```python
nonderiv_clean = nonderiv_clean.dropna(
    subset=['shares', 'shares_after']
)
```

```python
nonderiv_clean['direct_ownership'] = nonderiv_clean[
    'directOrIndirectOwnership'
].map({
    'D': 1,
    'I': 0
})
```

```python
nonderiv_clean.info()
```

```python
nonderiv_clean = nonderiv_clean.drop(
    columns=['directOrIndirectOwnership']
)
```

```python
nonderiv_clean.info()
```

```python
owner_sample.info()
```

```python
owner_clean = owner_sample[[
    'accessionNumber',
    'isDirector',
    'isOfficer',
    'isTenPercentOwner',
    'officerTitle'
]]
```

```python
role_cols = ['isDirector', 'isOfficer', 'isTenPercentOwner']

for col in role_cols:
    owner_clean[col] = owner_clean[col].fillna(0).astype(int)
```

```python
owner_clean['officerTitle'] = owner_clean['officerTitle'].fillna('Unknown')
```

```python
owner_clean.info()
```

```python
merged_df = nonderiv_clean.merge(
    owner_clean,
    on='accessionNumber',
    how='left'
)
```

```python
role_cols = ['isDirector', 'isOfficer', 'isTenPercentOwner']

merged_df[role_cols] = merged_df[role_cols].fillna(0)
merged_df['officerTitle'] = merged_df['officerTitle'].fillna('Unknown')
```

```python
merged_df.info()
```

```python
role_cols = ['isDirector', 'isOfficer', 'isTenPercentOwner']

merged_df[role_cols] = merged_df[role_cols].astype(int)
```

```python
merged_df['is_top_insider'] = (
    (merged_df['isOfficer'] == 1) |
    (merged_df['isDirector'] == 1) |
    (merged_df['isTenPercentOwner'] == 1)
).astype(int)
```

```python
merged_df['filing_transaction_gap'] = (
    merged_df['filingDate'].astype(str).str[:8].astype(int)
    - merged_df['transactionDate'].dt.strftime('%Y%m%d').astype(float)
)
```
